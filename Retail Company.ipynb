### Data Analysis

```python
# Datetime analysis

#using the .info(), this show that dtype of the date column is still object (categorical data). we need to change it to datetime

sales.info()
```

```python
# To convert the date column into pandas datatime object

sales["Date"] = pd.to_datetime(sales["Date"])

sales.info()
```

```python
# To Extract the Day, Month, Quarter, Year

sales["year"] = sales["Date"].dt.year
sales["month_name"] = sales["Date"].dt.month_name()
sales["month"] = sales["Date"].dt.month
sales["quarter"] = sales["Date"].dt.quarter
sales["day"] = sales["Date"].dt.day

sales.head()
```

```python
# To group Customers_Age

def age_group(x):
    if x <= 25:
        return "<=25 (Young Adult)"
    elif x <= 40:
        return "25 - 40 (Adult)"
    if x <= 50:
        return "41 - 50 (Old Adult)"
    else:
        return ">=51 (Elders)"

# Apply function to the data

sales['age_group'] = sales['Customer_Age'].apply(age_group)

sales.head()
    
```

```python
# To Calculate Cost, Revenue, Profit

sales['cost'] = sales['Quantity'] * sales['Unit_Cost']
sales['revenue'] = sales['Quantity'] * sales['Unit_Price']
sales['profit'] = sales['revenue'] - sales['cost']

sales.head()
```

```python
# To group Prifit/loss

def porl(x):
    if x >= 0:
        return 'profit'
    else:
        return 'loss'

# Apply function to the data

sales['profit_label'] = sales['profit'].apply(porl)

sales.head()
```
